```python
import urllib.request
import pandas # importa libreria occupa data analisys
import matplotlib as mpl
```

```python
dataset_url = "https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni.csv"

# lettura dataset da GitHub
urllib.request.urlretrieve(dataset_url, "dataset.cvs") # prende il file dalla url e lo scarica rinominandolo dataset.csv
```

```python
dataset = pandas.read_csv("dataset.cvs", parse_dates=['data'], index_col=['data'])
dataset.tail(50)
```

```python
emilia_romagna = dataset[dataset.denominazione_regione == 'Emilia-Romagna'] # cerca solo le righe della 
emilia_romagna[['denominazione_regione', 'totale_positivi', 'nuovi_positivi', 'deceduti']].tail()# filtra i dati e stampa solo i richiesti
```

```python
#@title Dimensione {run: "auto"}
dim = 125 #@param {type: "slider", min: 95, max: 200, step: 10}

with mpl.rc_context({'figure.dpi': 115, 'figure.facecolor': 'white'}):
  emilia_romagna.plot(y='totale_positivi', ylabel = 'totale positivi', \
                      grid=True, lw = 0.5, color='green')
```

```python
#@title Dimensione {run: "auto"}
dim = 105 #@param {type: "slider", min: 95, max: 200, step: 10}


with mpl.rc_context({'figure.dpi': dim, 'figure.facecolor': 'white'}):
  emilia_romagna.plot(y='totale_positivi_test_antigenico_rapido', \
                      title='Totale dei positivi al test antigenico rapido', \
                      ylabel = 'totale positivi test antigenico rapido', \
                      grid=True, lw = 2, color='orange', legend=True)
```
